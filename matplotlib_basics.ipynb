# Customizing Sizeof plots

```python
df['rest_type'].value_counts()
```

```python
df['rest_type'].value_counts().index
```

```python
df['rest_type'].value_counts().values
```

```python
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values)
```

```python
plt.figure(figsize=(20,12))
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values)
```

```python
plt.figure(figsize=(20,12))       #####to increase the chart in size-height and width
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values)

plt.gcf().autofmt_xdate()      ####To rotate the labels in the x-axis from horizontal to slanted form
```

```python
plt.figure(figsize=(20,12))
plt.bar(x=df['rest_type'].value_counts().index,
       height=df['rest_type'].value_counts().values)

plt.xticks(rotation='vertical')      ####to make the x-axis labels written in vertical form
plt.show()                           ####It get rid of unncessary writeups before the chart
```

```python
plt.scatter(x='rate',y='approx_cost(for two people)',data=df,c='g',s=10)       #####s=size and c=color
```

```python
plt.scatter(x='rate',y='approx_cost(for two people)',data=df,c='g',s=10)
plt.xlim(4.59,5.0)          #####used to set a limit or range of values to be displayed on the visual
plt.show()
```

```python
plt.scatter(x='rate',y='approx_cost(for two people)',data=df,c='g',s=10)
plt.xlim(4.59,5.0)
plt.ylim(1000,3000)
plt.show()
```

```python
location=df['location'].unique()
```

```python
plt.figure(figsize=(20,9))
for index,loc in enumerate(location,1):
    plt.subplot(3,2,index)
    data=df[df['location']==loc]
    plt.hist(data['rate'])
    plt.ylabel('count')
    plt.xlabel(loc)
```
