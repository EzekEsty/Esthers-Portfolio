#Creating Dictionaries, we use the curly brackets

```python
NOUN = {'Names': ['John', 'Kemi', 'Tomiwa'],
       'Things': ['Chair', 'Carpet','Book'],
       'Place': ['Abuja', 'Lagos', 'Ilorin'],
       'Animal': ['Pig', 'Lion', 'Tiger']}
```

```python
type(NOUN)
```

```python
#To pass it into a datframe
data_NOUN = pd.DataFrame(NOUN)
data_NOUN
```

```python
#To call a single column from the dataframe using the square bracket
data_NOUN['Things']
```

```python
#To call a single column from the dataframe using the dot
data_NOUN.Names
```

```python
#To check the datatype
type(data_NOUN.Names)
```

# Filtering Data

```python
#To get states with higher sales
highest_sales = (data['Sales'] > 300)
highest_sales
```

```python
data.loc[highest_sales]
```

```python
#To get states with highest sales with selected columns
data.loc[highest_sales, ['Product', 'Profit', 'Sales']]
```

```python
#To filter the states with specific products
data['Product'].unique()      ###to get the unique product names in the dataset
```

```python
#Create a variable named products
products = ['Green Tea', 'Earl Grey', 'Lemon']
```

```python
filter = data['Product'].isin(products)
```

```python
data.loc[filter]
```
