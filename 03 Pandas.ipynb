# Data Cleaning

```python
#To get the average salary per gender
# we need to clean the data by removing the dollar sign wich made the column a string

def cleanup_salary(row):      #####defining a new function
    salary = row['Salary'].replace('$', '')
    salary = float(salary)
    return salary

data['clean_salary'] = data.apply(cleanup_salary, axis=1)
```

```python
data.groupby('gender')['clean_salary'].mean()
```

```python
data.head()
```

# Deeper insight int the data

# Grouping Data with groupby()

```python
def female_salary(row):
    if row['gender'] == 'Female':
        return row['clean_salary']

data['Fem_Salary_Avg'] = data.apply(female_salary, axis=1)
```

```python
def male_salary(row):
    if row['gender'] == 'Male':
        return row['clean_salary']
    
data['Male_Salary_Avg'] = data.apply(male_salary, axis=1)
```

```python
#To group by job title

data.groupby('Job Title')['clean_salary', 'Fem_Salary_Avg', 'Male_Salary_Avg'].mean()
```

```python
#To group by Job title and city

data.groupby(['Job Title', 'City'])['clean_salary', 'Fem_Salary_Avg', 'Male_Salary_Avg'].mean()
```

Note: you can also use the median, min, max, sum

```python
#Task 2: Total bytes per domain split by male or female

#First define a function to calculate total female ytes and total male bytes

def female_bytes(row):
    if row['gender'] == 'Female':
        return row['bytes']
```

```python
#define a function to calculate total male bytes
def male_bytes(row):
    if row['gender'] == 'Male':
        return row['bytes']
```

```python
#Create new coulumns populated by the function
Url_data['female_bytes'] = Url_data.apply(female_bytes, axis=1)
Url_data['male_bytes'] = Url_data.apply(male_bytes, axis=1)
```

```python
Url_data.head()
```

```python
#group by url and sum male_bytes and female_bytes

Url_data.groupby('url')['male_bytes', 'female_bytes'].sum()
```

```python
#Task 3: Total bytes by gender and and country

Url_data.groupby(['gender', 'country'])['bytes'].sum().reset_index()
```

```python
#Task 4: Total bytes by country, url, and total male and female bytes

Url_data.groupby(['country', 'url'])['male_bytes', 'female_bytes'].sum()
```
