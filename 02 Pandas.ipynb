```python
import pandas as pd
```

```python
data = pd.read_csv('Data/Sales.csv')
```

```python
data.head()
```

#To access the columns, python does not recognize a space in column names. It should be repaced wit "underscore_" 

```python
data.columns = data.columns.str.replace(' ', '_')
```

```python
data.head(5)
```

```python
#To access a specific column
data.Product_Type
```

```python
#To rename a column
data.rename(columns = {'Marketing': 'Trade'})
```

```python
data.head(5)
```

#The change of column name is not yet effected into the data frame. Use the Inplace=True to get it changed 

```python
data.rename(columns = {'Marketing': 'Trade'}, inplace=True)
```

```python
data.head(5)
```

# To modify a value in a row

```python
#For example, to modify the value in the 3rd row of Total_Expenses

data.loc[2, 'Total_Expenses'] = 350
```

```python
data.head(5)
```

```python
#The 'at' attribute can also be used instaed of the loc
data.at[2, 'Total_Expenses'] = 500
```

```python
data.head()
```

# Adding/Removig rows and columns from pandas data frame

```python
#To add two columns
new_col = data['Total_Expenses'] + data['Sales']
```

```python
new_col
```

```python
#To add the new column to the data frame
data['New_Column'] = new_col
data
```

```python
#To remove one or morre columns

data.drop(columns=['Product', 'Market'])
```

#To get it effected in the original data frame, use the "inplace=True"

```python
#To remove a row

data.drop(index = 0)
```

# Data Sorting in Pandas

```python
#To sort the data by a single column e.g sales

data.sort_values(by='Sales')
```

```python
#To sort using mulltiple columns and in descending order

data.sort_values(by=['Sales','Trade'], ascending=False)
```

```python
#To sort by index or row-wise

data.sort_index(ascending=False)
```

```python
#To sort a string, i.e aphabetically

data['Market'].sort_values()
```

# Cleaning Data

```python
#To check for missing value(s)

data.isnull().sum()
```

```python
#To clean the data, one way is to drop the affected row(s)

data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

# Saving data in different format

```python
#To save in csv format

data.to_csv('New_Sales.csv')
```
